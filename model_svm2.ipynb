```python
import json
import numpy as np
import pandas as pd
```

```python
dataJSON  = []

with open("/content/drive/MyDrive/Colab Notebooks/apt/df_pp_word2vec.json", 'r') as f:
  dataJSON=json.load(f)
  df_pp_fasttext = pd.DataFrame(dataJSON)

df_pp_fasttext["allVectors"] = df_pp_fasttext["query_vector"] + df_pp_fasttext["product_vector"] 
train = np.array([np.array(list(np.float_(x))) for x in df_pp_fasttext["allVectors"]])

label = df_pp_fasttext["median_relevance"] - 1
```

```python
print(train.shape)
print(label.shape)
```

## SVM

### SVM NAD KONKATENIRANIM VEKTORIMA UPITA I PROIZVODA BEZ BROJEVA

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, label, test_size=0.33, random_state=42)
model = SVC(kernel = 'rbf', gamma = 0.4).fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
conf = confusion_matrix(y_test, prediction)
```

```python
#sa brojevima
print(np.sum(np.diag(conf))/len(prediction))
```

```python
#bez brojeva
print(np.sum(np.diag(conf))/len(prediction))
```
